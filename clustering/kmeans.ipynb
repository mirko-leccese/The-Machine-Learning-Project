# Introduction to K-Means

**K-Means** is one of the most widely used *clustering* techniques, i.e. an unsupervised method for identifying groups or *clusters* of data points in a given multi-dimensional space. In this notebook, we provide a mathematical introduction to K-Means, as well as an implementation from scratch in Python, albeit not optimized for large dataset. The notebook is organized as follows:
- **Mathematical Formulation**: this section undertakes the mathematical foundation of the clustering problem and its solution with the naive K-Means approach;
- **Python Implementation**: in this section, we present a Python class with the basic implementation of naive K-Means, and methods to compute additional relevant metrics such as the *silhoutte*;
- **Scikit-Learn Implementation**: in this section, we illustrate the use of the Scikit-Learn package to perform K-Means on dataset of increasing complexity

## Mathematical Formulation

Let ${\mathbf{x}_1, ..., \mathbf{x}_N}$ be a data set consisting of $N$ observations of a random $D$-dimensional Euclidean variable $\mathbf{x}$. Our goal is to partition the data set into $K$ clusters, where we suppose for the moment that the value of $K$ is given. Intuitively, we can think of a cluster as a group of data points whose inter-point distances are small if compared to the distances to points outside that cluster. Let then define $K$ vectors $\mathbf{\mu}_k$ representing the *centres* of the clusters. The problem translates into finding the set of vectors $\{\mathbf{\mu}_k\}$, as well as the assignment of data points to clusters, such that *the sum of squares of the distances of each data point to its closest vector $\mathbf{\mu}_k$ is a minimum.*

As it is standard in many machine learning techniques, we should define a *cost function* $J$ and then compute its derivatives, to find the values of model parameters which minimizes such cost function. To this aim, let introduce a *binary indicator variable* $\chi_{nk} \in \{0, 1\}$ where $k=1, 2, ..., K$ which indicates which cluster a data point $\mathbf{x}_n$ is assigned to. So, if a point belong to cluster $k=j$, then $\chi_{nk} = \delta_{kj}$. This is at times known as the **1-of-K coding scheme**. We define the cost function $J$ (also known as *distorsion measure*) as:

$$ J = \sum_{n=1}^N \sum_{k=1}^K \chi_{nk} || \mathbf{x}_n - \mathbf{\mu}_k ||^2 $$

i.e. it is the sum of the squares of the distances ($L^2$ norm) of each data point to its assigned cluster centers. The objective is to find the set of values $\{\chi_{nk}\}$ and $\{\mathbf{\mu}_k\}$ which minizes $J$. The optimization can be performed in a two-step iterative procedure:
- (1) We choose some initial values for $\{\mathbf{\mu}_k\}$. Then, we keep them fixed and minimize $J$ w.r.t. $\chi_{nk}$;
- (2) We minimize $J$ w.r.t $\{\mathbf{\mu}_k\}$, keeping $\chi_{nk}$ fixed.
The first optimization is trivial since $J$ is a linear function of $\chi_{nk}$. Indeed, we can choose the $\chi_{nk}$ separately for each $n$ to be 1 for whichever value of $k$ giving the minimum value of the $L^2$ norm $||\mathbf{x}_n - \mathbf{\mu}_k ||^2$. In other words, we simply assign the point $\mathbf{x}_n$ to the closest cluster centre:
$$ \chi_{nk} = \begin{cases}
    1 \ \ \ \text{if} \ k = \arg \min_j  || \mathbf{x}_n - \mathbf{\mu}_j ||^2 \\
    0 \ \ \ \text{otherwise}
  \end{cases}
$$
The second step consists instead of keeping $\chi_{nk}$ fixed and minimize the objective function $J$ w.r.t. $\mathbf{\mu}_k$ . Hence,

$$
\begin{align}
\frac{\partial J}{\partial \mathbf{\mu}_k} &= \frac{\partial}{\partial \mathbf{\mu}_k} \sum_{n=1}^N \sum_{k=1}^K \chi_{nk} || \mathbf{x}_n - \mathbf{\mu}_k ||^2 \\
&=  2 \sum_{n=1}^N \chi_{nk} (\mathbf{x}_k - \mathbf{\mu}_k) = 0
\end{align}
$$
where the second summation over $k$ has been removed since in this step each point has been assigned to the closest cluster centre. We can solve the above equation for $\mathbf{\mu}_k$ to give:
$$
\mathbf{\mu}_k = \frac{\sum_n \chi_{nk} \mathbf{x}_n}{\sum_n \chi_{nk}}
$$
The above equation has a very simple and elegant interpretation. The denominator of the expression represents the number of points assigned to cluster $k$; in the numerator, the expression $\chi_{nk} \mathbf{x}_n$ will select just data points belonging to cluster $k$ since $\chi_{nk}$ is $\delta_{jk}$. Therefore, $\mathbf{\mu}_k$ are simply *set equal to the mean of all of the data points $\mathbf{x}_n$ assigned to cluster $k$*. For this reason, this procedure is referred to as the **K-Means algorithm**. 

These two phases of re-assigning data points to the closest cluster centre and re-computing the cluster means can be repeated until no change in the assignment occurs. Since each step reduced the objective cost function, convergence is assured, although the procedure may converge to a local minimum rather than a global minimum of $J$. 

## Python Implementation

In this section, we provide a basic implementation of the *K-Means* algorithm in Python. Altough we make extensive use of the `numpy` package which allow fast vector/matrix operations, the implementation given here is not properly optimized for large datasets and therefore is not intended for real-life applications.

We now implement the *K-Means* algorithm as a Python class. The idea is to use the class in the following way:
- the user instantiate the class, passing basic parameters such as the number of clusters (`num_clusters`), the maximal number of optimization iterations (`max_iters`) and a tolerance for convergence (`tol`);
- hence, the user calls the `fit` method of the class, which should accept the data set as main argument;
- the method returns the optimal cluster assignments, which can then be plotted to visualize clusters.

```python
def timer(func):
    '''
    A decorator that prints how long a function took to run.
    : params func (callable): The function being decorated
    :returns callable: The decorated function.
    '''
    
    def wrapper(*args, **kwargs):
        # When wrapper() is called, get the current time
        t_start = time.time()
        
        # Call the decorated function and store the result.
        result = func(*args, **kwargs)
        
        # Get the total time it took to run and print it
        t_total = time.time() - t_start
        
        print('{} took {}s'.format(func.__name__, t_total))
        return result		
    return wrapper
```

```python
class naiveKMeans:
    def __init__(self, num_clusters: int, max_iters: int, tol: float, num_steps: int = 10) -> Tuple[np.array, np.array]:
        self.num_clusters = num_clusters
        self.max_iters = max_iters
        self.num_steps = num_steps
        self.tol = tol
        self._J = np.array([])

    def __initialize_centres(self, data: np.array) -> np.array:
        '''
        Initialization function that randomly select cluster centers from data.

        :params data: the dataset as NumPy array
        :returns: the randomly selected cluster centres
        '''
        #np.random.shuffle(data)
        #return data[:self.num_clusters]
        return data[np.random.choice(len(data), self.num_clusters, replace=False)]


    def __assign_clusters(self, data: np.array, centres: np.array) ->  Tuple[np.array, np.array]:
        '''
        Function that assign each point to the cluster with the closest cluster centers (min of Euclidean distance).

        :params data: the dataset as NumPy array
        :params centres: the numpy array of centres
        :returns: a tuple whose elements are the numpy array of cluster assignements and the numpy array of distances to the clusters
        '''

        distances = [np.linalg.norm(centres - point, axis=1) for point in data]
        clusters = [np.argmin(x) for x in distances]

        return np.array(clusters), np.array(distances)

    def __update_centres(self, data: np.array, clusters: np.array) -> np.array:
        '''
        Function that update the cluster centres computing the mean of data points previously assigned to cluster k

        :params data: the dataset as a Numpy array
        :params clusters: the previous cluster assignments as a numpy array
        :returns: the numpy array of new centres
        '''

        # Taking points belonging to a given cluster
        # clusters is a np.array of the "data" dimension with cluster labels (0,1, etc) for each point; clusters == i returns an
        # np.array of true/false values where true if point belong to cluster i. Storing everything in a list
        clusters_points = [data[clusters == i] for i in range(self.num_clusters)]

        # computing the mean (centroid) of data points belonging to cluster i
        new_centres = [np.mean(cluster_points, axis=0) for cluster_points in clusters_points]

        return np.array(new_centres)

    
    def _single_fit(self, data: np.array) -> Tuple[np.array, np.array, np.array]:
        '''
        The main method that "fits" the algorithm to the passed dataset.

        :params data: the dataset as a Numpy array
        :returns: a tuple whose elements are the optimal cluster assignements and the cluster centres
        '''
        # Initialize cluster centers
        centres = self.__initialize_centres(data)

        J = []
        for iter in range(self.max_iters):
            # Assign points to the closest cluster
            clusters, distances = self.__assign_clusters(data, centres)

            # Appending the cost function after assigning points to the cluster
            J.append(sum([distance[clusters[i]] for i, distance in enumerate(distances)]))
    
            # Update cluster centers
            new_centres = self.__update_centres(data, clusters)

            if iter != 0:
                if abs(J[iter]-J[iter-1]) <= self.tol:
                    break
            # Check for convergence
            #if np.allclose(centers, new_centers):
            #    break
            
            centres = new_centres
        
        return clusters, centres, np.array(J)


    @timer
    def fit(self, data: np.array) -> Tuple[np.array, np.array]:

        runs = [self._single_fit(data) for step in range(self.num_steps)]
        
        # Find min J
        J_index = 2
        min_Js = [np.min(run[J_index]) for run in runs]

        # find the min run
        min_run_index = np.argmin(np.array(min_Js))

        self._J = runs[min_run_index][J_index]
        
        return runs[min_run_index][:J_index]
        
    def __compute_distance_matrix(self, data: np.array) -> np.array:
        '''
        Method to compute the distance matrix, i.e. a matrix whose element ij represent the distance between point i and j,
        i.e. in this case the L^2 norm of the vector (x_i - x_j)

        :params data: the dataset as a numpy array
        :returns: the distance matrix as numpy array (here an NxN matrix where N is the number of points)
        '''
        return np.linalg.norm(data[:, np.newaxis]-data, axis=2)

    def __compute_mean_intracluster_distance(self, dist_matrix: np.array, i: int, clusters: np.array) -> float:
        '''
        Method to compute the mean intracluster distance for a given point, i.e. the mean of the distances between point i and each point
        belonging to the same cluster

        :params dist_matrix: the distance matrix
        :params i: the index of the point
        :params clusters: the cluster assignments
        :returns: the mean intracluster distance
        '''
        return np.mean(dist_matrix[i, clusters == clusters[i]])

    def __compute_dissimilarity(self, dist_matrix: np.array, i: int, clusters: np.array) -> float:
        '''
        Method to compute the dissimilarity measure for a given point, i.e. the mean of the distances between point i and each point
        NOT belonging to the same cluster

        :params dist_matrix: the distance matrix
        :params i: the index of the point
        :params clusters: the cluster assignments
        :returns: the dissimilarity measure
        '''
        
        return np.mean(dist_matrix[i, clusters != clusters[i]])

    def __point_silhoutte(self, data: np.array, clusters: np.array) -> np.array:
        '''
        Method to compute the silhoutte score for all points, which is a measure of the "goodness" of the current cluster assignment
        
        :params data: the data as a numpy array
        :params clusters: the cluster assignments
        :returns: a numpy array with point silhoutte scores
        '''
        n = len(data)

        dist_matrix = self.__compute_distance_matrix(data)
        
        mean_intracluster_array = np.array([self.__compute_mean_intracluster_distance(dist_matrix, i, clusters) for i in range(n)])
        dissimilarity_array = np.array([self.__compute_dissimilarity(dist_matrix, i, clusters) for i in range(n)])

        return np.where(
            mean_intracluster_array < dissimilarity_array, 
            1 - mean_intracluster_array/dissimilarity_array,
            np.where(
                mean_intracluster_array == dissimilarity_array,
                0,
                dissimilarity_array/mean_intracluster_array -1
            )
        )

    def get_silhoutte(self, data: np.array, clusters: np.array) -> float:
        '''
        Method to compute the clustering silhoutte, i.e. the mean of point silhoutte scores. 

        :params data: the dataset as a numpy array
        :params clusters: the cluster assignments
        :returns: the silhoutte score
        '''
        return np.mean(self.__point_silhoutte(data, clusters))

    @staticmethod
    def plot_clusters(data, clusters, centers):    
        fig, ax = plt.subplots(figsize=(6,6))
        
        # Plot the data points with consistent colors for each cluster
        ax.scatter(data[:, 0], data[:, 1], c=clusters, cmap="tab10", marker='o', alpha=0.7)
        
        # Plot the cluster centers with fixed color (red)
        ax.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100)
        
        # Add title, labels, and colorbar
        ax.set_title('naive K-means Clustering')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        
        # Show the plot
        plt.show()
```

We now apply the implemented K-Means to the input dataset:

```python
# create an instance of naiveKMeans class
kmeans = naiveKMeans(num_clusters = 10, max_iters = 10000, tol=1e-8, num_steps=50)
```

```python
# call the fit method
clusters, centers = kmeans.fit(X)
```

```python
sil_k = []
for k in range(2, 12):
    kmeans_k = naiveKMeans(num_clusters = k, max_iters= 10000, tol=1e-8, num_steps=50)
    clusters_k, _ = kmeans_k.fit(X)

    sil_k.append(kmeans_k.get_silhoutte(X, clusters_k))
```
