```python
# Generate 3 random clusters with 100 points in each
num_points = 600
num_clusters = 6
X, _ = make_blobs(n_samples=num_points*num_clusters, centers=num_clusters, cluster_std=1.1)
```

```python
# Create a figure and axis object
fig, ax = plt.subplots(figsize=(6,6))

# Plot the scatter plot using ax.scatter()
ax.scatter(X[:, 0], X[:, 1], color='royalblue', marker='o')

# Set the title and labels
ax.set_title('Random Scatter Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Show the grid
ax.grid(True)
```

```python
class naiveKMeans:
    def __init__(self, num_clusters: int, max_iters: int, tol: float) -> Tuple[np.array, np.array]:
        self.num_clusters = num_clusters
        self.max_iters = max_iters
        self.tol = tol
        self.J_ = []

    def __initialize_centers(self, data):
        '''
        Initialization function that randomly select cluster centers from data
        '''
        np.random.shuffle(data)
        return data[:num_clusters]


    def __assign_clusters(self, data, centers):
        '''
        Function that assign each point to the cluster with the closest cluster centers (min of Euclidean distance)
        '''

        distances = [np.linalg.norm(centers - point, axis=1) for point in data]
        clusters = [np.argmin(x) for x in distances]

        return np.array(clusters), np.array(distances)

    def __update_centers_old(self, data, clusters):
        new_centers = []
        for i in range(self.num_clusters):
            # Taking points belonging to a given cluster
            # clusters is a np.array of the "data" dimension with cluster labels (0,1, etc) for each point; clusters == i returns an
            # np.array of true/false values where true if point belong to cluster i
            cluster_points = data[clusters == i]

            # computing the mean (centroid) of data points belonging to cluster i
            new_center = np.mean(cluster_points, axis=0)

            new_centers.append(new_center)

        return np.array(new_centers)

    def __update_centers(self, data, clusters):

        # Taking points belonging to a given cluster
        # clusters is a np.array of the "data" dimension with cluster labels (0,1, etc) for each point; clusters == i returns an
        # np.array of true/false values where true if point belong to cluster i. Storing everything in a list
        clusters_points = [data[clusters == i] for i in range(self.num_clusters)]

        # computing the mean (centroid) of data points belonging to cluster i
        new_centers = [np.mean(cluster_points, axis=0) for cluster_points in clusters_points]

        return np.array(new_centers)
    
    def fit(self, data):
        # Initialize cluster centers
        centers = self.__initialize_centers(data)
    
        for iter in range(self.max_iters):
            # Assign points to the closest cluster
            clusters, distances = self.__assign_clusters(data, centers)

            # Appending the cost function after assigning points to the cluster
            self.J_.append(sum([distance[clusters[i]] for i, distance in enumerate(distances)]))
    
            # Update cluster centers
            new_centers = self.__update_centers(data, clusters)

            if iter != 0:
                if abs(self.J_[iter]-self.J_[iter-1]) <= self.tol:
                    break
            # Check for convergence
            #if np.allclose(centers, new_centers):
            #    break
            
            centers = new_centers
        
        return clusters, centers

    def __compute_distance_matrix(self, data: np.array) -> np.array:
        return np.linalg.norm(data[:, np.newaxis]-data, axis=2)

    def __compute_mean_intracluster_distance(self, dist_matrix, i, clusters):
        return np.mean([dist_matrix[i], clusters == clusters[i]])

    def __compute_dissimilarity(self, dist_matrix, i, clusters):
        return np.mean([dist_matrix[i], clusters != clusters[i]])

    def __point_silhoutte(self, data, clusters):
        n = len(data)

        dist_matrix = self.__compute_distance_matrix(data)
        
        mean_intracluster_array = np.array([self.__compute_mean_intracluster_distance(dist_matrix, i, clusters) for i in range(n)])
        dissimilarity_array = np.array([self.__compute_dissimilarity(dist_matrix, i, clusters) for i in range(n)])

        return np.where(
            mean_intracluster_array < dissimilarity_array, 
            1 - mean_intracluster_array/dissimilarity_array,
            np.where(
                mean_intracluster_array == dissimilarity_array,
                0,
                dissimilarity_array/mean_intracluster_array -1
            )
        )

    def get_silhoutte(self, data, clusters):
        return np.mean(self.__point_silhoutte(data, clusters))

    @staticmethod
    def plot_clusters(data, clusters, centers):    
        fig, ax = plt.subplots(figsize=(6,6))
        
        # Plot the data points with consistent colors for each cluster
        ax.scatter(data[:, 0], data[:, 1], c=clusters, cmap="tab10", marker='o', alpha=0.7)
        
        # Plot the cluster centers with fixed color (red)
        ax.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100)
        
        # Add title, labels, and colorbar
        ax.set_title('naive K-means Clustering')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        
        # Show the plot
        plt.show()
```

```python
kmeans = naiveKMeans(num_clusters = 6, max_iters = 50000, tol=1e-12)
```

```python
clusters, centers = kmeans.fit(X)
```

```python
kmeans.plot_clusters(X, clusters, centers)
```

```python
cost_function = kmeans.J_
cost_function
```

```python
fig, ax = plt.subplots(figsize=(6,6))

# Plot the data points with consistent colors for each cluster
ax.plot(cost_function, linewidth=2.0, marker='o', alpha=0.7)

# set xticks
ax.set_xticks(range(0,len(cost_function)))


# Add title, labels, and colorbar
ax.set_title('naive K-means Cost Function')
ax.set_xlabel('N iterations')
ax.set_ylabel('J')

# Show the plot
plt.show()
```

```python
sil_k = []
for k in range(2, 8):
    kmeans_k = naiveKMeans(num_clusters = k, max_iters= 10000, tol=1e-12)
    clusters_k, _ = kmeans_k.fit(X)

    sil_k.append(kmeans_k.get_silhoutte(X, clusters_k))
```
